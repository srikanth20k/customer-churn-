```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
import pickle
```

```python
df = pd.read_csv(r'C:\Users\SRIKANTH ADIPIREDDY\Desktop\New folder\DeepLearning\churn_project\Churn_Modelling.csv')
```

```python
df.head()
```

### Preprocessing
* drop irrelevent columns

```python
data = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
data.head()
```

```python
data
```

### Label Encoding

```python
lb = LabelEncoder()

data['Gender']= lb.fit_transform(data['Gender'])
```

```python
data
```

### Onehot Encoding

```python
onehot = OneHotEncoder(sparse_output=False)

one_data = onehot.fit_transform(data[['Geography']])
```

```python
one_data
```

```python
onehot.get_feature_names_out(['Geography'])
```

```python
data_onehot = pd.DataFrame(one_data, columns=onehot.get_feature_names_out(['Geography']))
```

```python
data_onehot
```

```python
data = pd.concat([data, data_onehot], axis=1).drop(['Geography'], axis=1)
```

```python
data.head()
```

```python
### Label Encoding
with open('label_encoder.pkl', 'wb') as f:
     pickle.dump(lb, f)
    
### Onehot Encoding
with open('onehot_encoder.pkl', 'wb') as f:
    pickle.dump(onehot, f)
```

### splitting the data into training and test data
* Divide the dataset into indepent and dependent features
* spliting the data indepent and dependent into train ,test split 

```python
x = data.drop('Exited', axis=1)
y = data['Exited']

#### splitting the data into training and test data
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

## scaling the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## saving the scaler
with open('scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
```

```python
X_train.shape[1]
```

### ANN Impementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
```

```python
X_train.shape
```

```python
model = Sequential([
    Dense(units=64, activation='relu', input_shape=(X_train.shape[1],)), #input layer
    Dense(units=32, activation='relu'),#hidden layer
    Dense(units=1, activation='sigmoid')#output layer
])
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
import datetime #for tensorboard callback
log_dir = "logs/fit/"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
### early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# train the model
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=100,
                     callbacks=[early_stopping, tensorboard_callback])
```

```python
model.save('churn_model.h5')
```
